```python
import matplotlib.pyplot as plt

heures = [1, 2, 3, 4, 5, 6]
tas = [125, 120, 110, 100, 95, 88]
fc = [80, 85, 95, 105, 115, 125]
du = [100, 80, 50, 30, 20, 10]
fr = [14, 16, 18, 20, 22, 25]
hb = [12.5, 12.1, 11.5, 10.8, 9.9, 9.0]
spo2 = [97, 97, 96, 95, 95, 94]

# 1. Création de la Figure (2 lignes, 3 colonnes)
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))

# --- LIGNE 1 (axes[0, x]) ---

# 1A: T.A. Systolique (Haut-Gauche)
axes[0, 0].plot(heures, tas, color='blue', marker='o')
axes[0, 0].set_title('T.A. Systolique (mmHg)')
axes[0, 0].set_xlabel("Heures")

# 1B: Fréquence Cardiaque (Haut-Centre)
axes[0, 1].plot(heures, fc, color='red', marker='o')
axes[0, 1].set_title('Fréquence Cardiaque (bpm)')
axes[0, 1].set_xlabel("Heures")

# 1C: Débit Urinaire (Haut-Droite)
axes[0, 2].plot(heures, du, color='orange', marker='o')
axes[0, 2].set_title('Débit Urinaire (mL/h)')
axes[0, 2].set_xlabel("Heures")

# --- LIGNE 2 (axes[1, x]) ---

# 2A: Fréquence Respiratoire (Bas-Gauche)
axes[1, 0].plot(heures, fr, color='purple', marker='o')
axes[1, 0].set_title('Fréquence Respiratoire (c/min)')
axes[1, 0].set_xlabel("Heures")

# 2B: Hémoglobine (Bas-Centre)
axes[1, 1].plot(heures, hb, color='brown', marker='o')
axes[1, 1].set_title("Taux d'Hémoglobine (g/dL)")
axes[1, 1].set_xlabel("Heures")

# 2C: SpO2 (Bas-Droite)
axes[1, 2].plot(heures, spo2, color='green', marker='o')
axes[1, 2].set_title('Saturation O₂ (%)')
axes[1, 2].set_xlabel("Heures")

# 3. Ajustement de l'espace (obligatoire pour 6 graphiques !)
fig.suptitle('Tableau de Bord de Surveillance Post-Opératoire', fontsize=16)
fig.subplots_adjust(wspace=0.3, hspace=0.4)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
c = 299792458
m = np.linspace(0,10,11)
E = m*c**2


fig, ax = plt.subplots()
ax.plot(m,E,color= 'red', lw=5)
ax.set_xlabel("masse en gramme")
ax.set_ylabel("énergie en joules")
ax.set_title("E=mc^2")
plt.yscale('log')
plt.grid(which= 'both', axis= 'y')
plt.show()
```

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']
july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]


fig, ax1 = plt.subplots(figsize=(12,8))

ax1.plot(labels,july16_2007, lw=2, color="blue")
ax1.set_ylabel("2007", fontsize=18, color="blue")

ax1.spines['left'].set_color('blue')
ax1.spines['left'].set_linewidth(4)

for label in ax1.get_yticklabels():
    label.set_color("blue")
plt.yticks(fontsize=15)

ax2 = ax1.twinx()
ax2.plot(labels,july16_2020, lw=2, color="red")
ax2.set_ylabel("2020", fontsize=18, color="red")

ax2.spines['right'].set_color('red')
ax2.spines['right'].set_linewidth(4)

for label in ax2.get_yticklabels():
    label.set_color("red")

ax1.set_title("Courbes de rendement du 16 juillet");
plt.yticks(fontsize=15);
```
